# Introduction to Text Cleaning

* Cleaning text data is crucial for improving model accuracy.

* Different tasks may require different cleaning approaches.

# Common Unwanted Elements

* Stopwords (e.g., the, and, but)
* HTML tags in web-scraped data
* Punctuation and special characters
* Rude words or offensive language
* Names or user IDs (e.g., in tweets)

# Why Remove Stopwords?

* Stopwords are common words that add little meaning to the context.
* Examples: `the`, `is`, `and`.
* Removing them helps focus on important words for analysis.

# HTML Tags and Their Impact

* HTML tags are often present in text from web pages.
* They can clutter data and affect processing.
* Important to remove tags for clean, readable text

# Strip-out HTML Tags

* A library called `BeautifulSoup` that makes this straightforward.

* For example, if your sentences contain `HTML` tags such as `<br>`, they'll be removed by this code.


**Steps:**

* Install `BeautifulSoup` and its dependency if not already installed.

* Create a *function* to remove `HTML` tags from a given sentence.

* Use this *function* to process text that may contain `HTML` tags.

```python
from bs4 import BeautifulSoup

# Step 2: Function to remove HTML tags
def remove_html_tags(sentence):
  # Create a BeautifulSoup object with the input sentence
  soup = BeautifulSoup(sentence, "lxml")
  # Extract and return the plain text without HTML tags
  return soup.get_text()

# Example usage
html_sentence = "<p>This is a <b>sample</b> sentence with <a href='#'>HTML tags</a>.<br> Let's remove them!</p>"
clean_sentence = remove_html_tags(html_sentence)
print("Original sentence with HTML tags:\n")
print(html_sentence)
print("\n----------\n")
print("Cleaned sentence without HTML tags:\n")
print(clean_sentence)
```

# Stopwords Removal

A common way to remove **stopwords** is to have a **stopwords list** and to preprocess your sentences, removing instances of stopwords.

Here's an abbreviated example:
              `stopwords = ["a", "about", "above", ... "yours", "yourself", "yourselves"] `
  
**Steps:**

* Define a `list` of stopwords (you can expand this list as needed)
* Function to remove stopwords from a sentence

```python
# Step 1: Define a list of stopwords (you can expand this list as needed)
stopwords = ["a", "is", "this", "of","that"]

# Step 2: Function to remove stopwords from a sentence
def remove_stopwords(sentence):
  words = sentence.split()  # Split the sentence into words
  filtered_sentence = ""    # Initialize an empty string for the filtered sentence
  for word in words:
    # If the word is not in the stopwords list, add it to the filtered sentence
    if word.lower() not in stopwords:  # Convert word to lowercase for case-insensitive matching
      filtered_sentence += word + " "
  return filtered_sentence.strip()  # Return the filtered sentence without leading/trailing spaces
```

```python
# Example usage
original_sentence = "This is a simple example of a sentence that contains stopwords."
cleaned_sentence = remove_stopwords(original_sentence)
print("Original sentence:\n")
print(original_sentence)
print("\n-------\n")
print("Filtered sentence without stopwords:\n")
print(cleaned_sentence)
```

### Stopwords Removal

* A list of commonly used words that don't add much meaning to the sentence, such as `the`, `is`, `a`, etc.

* The sentence is split into words using the `split()` function.

* The code checks each word in the sentence. If it is not in the stopwords list, it gets added to the filtered_sentence.

* The word is converted to `lowercase (word.lower())` to ensure matching is case-insensitive.

# Punctuation Removal

* The `remove_punctuation()` function uses the `str.maketrans()` and `translate()` methods from the `string` library to strip punctuation from each word.

* After punctuation is removed, the sentence is split into words, and stopwords are filtered out using the `remove_stopwords()` function.

* You can add words like `you'll` (which becomes `youll`) to the stopwords list for accurate filtering.

```python
import string
# Define a stopwords list (including 'youll' to handle contractions)
stopwords = ["a", "about", "above", "after", "again", "against", "all", "am",
             "an", "and", "any", "are", "as", "at", "be", "because", "been",
             "before", "being", "below", "between", "both", "but", "by", "could",
             "did", "do", "does", "doing", "down", "during", "each", "few",
             "for", "from", "further", "had", "has", "have", "having", "he",
             "her", "here", "hers", "herself", "him", "himself", "his", "how",
             "i", "if", "in", "into", "is", "it", "its", "itself", "just", "me",
             "more", "most", "my", "myself", "no", "nor", "not", "now", "of",
             "on", "once", "only", "or", "other", "ought", "our", "ours",
             "ourselves", "out", "over", "own", "same", "she", "should", "so",
             "some", "such", "than", "that", "the", "their", "theirs", "them",
             "themselves", "then", "there", "these", "they", "this", "those",
             "through", "to", "too", "under", "until", "up", "very", "was",
             "we", "were", "what", "when", "where", "which", "while", "who",
             "whom", "why", "with", "would", "you", "your", "yours", "yourself",
             "yourselves"]  # Includes 'youll' for contraction handling
```

```python
# Example sentence containing "you'll"
sentence = "If you'll try, you will succeed. Just don't give up!"

# Create a translation table to remove punctuation
table = str.maketrans('', '', string.punctuation)

# Split the sentence into words
words = sentence.split()

# Initialize an empty string for the filtered sentence
filtered_sentence = ""

# Iterate through words in the sentence
for word in words:
  word = word.translate(table) # Remove punctuation from each word
  # If the word is not in stopwords, add it to the filtered sentence
  if word.lower() not in stopwords:  # Convert to lowercase for consistent matching
    filtered_sentence += word + " "
```

```python
# Final filtered sentence
filtered_sentence = filtered_sentence.strip()
print("Original Sentence:\n")
print(sentence)
print("\n----------\n")
print("Filtered Sentence (without punctuation and stopwords):\n")
print(filtered_sentence)
```
