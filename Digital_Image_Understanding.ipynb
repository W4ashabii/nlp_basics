### Measure pixel intensity values

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
grayscale_image = cv2.imread('downloaded_image.png', cv2.IMREAD_GRAYSCALE)
```

```python
# Display the image
plt.imshow(grayscale_image, cmap='gray')
plt.title('Grayscale Bird Image')
plt.axis('off')
plt.show()
```

```python
# Calculate intensity statistics
mean_intensity = np.mean(grayscale_image)
max_intensity = np.max(grayscale_image)
min_intensity = np.min(grayscale_image)
print(f"Mean Intensity: {mean_intensity}")
print(f"Max Intensity: {max_intensity}")
print(f"Min Intensity: {min_intensity}")
```

```python
# Display a histogram of the intensity values
plt.hist(grayscale_image.ravel(), bins=256, range=(0, 256), color='black')
plt.title('Intensity Distribution')
plt.xlabel('Intensity Value')
plt.ylabel('Frequency')
plt.show()
```

**How Intensity Aids Diagnostics**

In medical imaging, pixel intensity often represents tissue density or the presence of certain materials (like bone or fluid).

For example: High intensity (whiter regions) in X-rays or CT scans might indicate denser structures like bone.

Low intensity (darker regions) might represent soft tissue or areas of low density.

This intensity information helps radiologists or automated systems detect abnormalities, such as tumors, fractures, or lesions, which appear as distinct intensity patterns.

## Identify RGB value from a particular pixel

```python
image_original = cv2.imread('/content/drive/MyDrive/Sample_Images/BIRD_COLOR.jpeg')
cv2_imshow(image_original)
```

## Resize the image

```python
# Resize: Scale by a factor
scale_factor = 0.2  # 20% of original
image2 = cv2.resize(image_original, None, fx=scale_factor, fy=scale_factor, interpolation=cv2.INTER_AREA)

cv2_imshow(image2)
```

```python
# Define the pixel coordinates you want to check
pixel_coordinates = [(100, 100), (200, 200), (300, 300)]

for x, y in pixel_coordinates:
  # Get the RGB values at the specified pixel
   (b, g, r) = image2[y, x] # Note: y comes first, then x

   # Print the RGB values
   print(f"Pixel at ({x}, {y}): R={r}, G={g}, B={b}")
```

## Channel Separation in RGB Images

Each color channel shows different details of an image.

Display Red, Green, and Blue channels separately.

```python
b, g, r = cv2.split(image2) # Separate color channels

# Create zero-filled images for each channel
zeros = np.zeros(image2.shape[:2], dtype="uint8")

# Merge channels back into color images, keeping only the desired channel
red_img = cv2.merge([zeros, zeros, r])
green_img = cv2.merge([zeros, g, zeros])
blue_img = cv2.merge([b, zeros, zeros])
```

```python
cv2_imshow(red_img)
```

```python
cv2_imshow(green_img)
```

```python
cv2_imshow(blue_img)
```

## Understanding Color Depth

24-bit RGB images have 8 bits per channel (millions of colors).

Reduced color depth affects image quality.

Code to reduce color depth is as follows:

```python
# Reduce color depth to 2 bits (4 colors per channel)
reduced_image = (image2 // 64) * 64
reduced_image = reduced_image.astype(np.uint8)
# Display the reduced image
# cv2_imshow(image2)
cv2_imshow(reduced_image)
```

## Activity - manipulating RGB Values

**Task:** Change RGB values of pixels.

**Discussion:** How does changing each channel affect the image?



```python
# Set the red channel to a specific value (e.g., 100)
image2[:, :, 2] = 0 #B=0,G=1,R=2

# Display the modified image
cv2_imshow(image2)
```

## Plot Histogram for RGB Channel

```python
# Calculate histogram for each color channel
colors = ('b', 'g', 'r') # OpenCV uses BGR order

for i, color in enumerate(colors):
  hist = cv2.calcHist([image2], [i], None, [256], [0, 256])
  plt.plot(hist, color=color)
  plt.xlim([0, 256])

  # Show the plot
  plt.title('Color Histogram')
  plt.xlabel('Pixel Value')
  plt.ylabel('Frequency')
  plt.show()
```

## Activity - brightness Adjustment

**Task:** Use code to brighten an image.

**Observation:** Impact on darker areas.

```python
def increase_brightness(image, value=100):
  return cv2.add(image, value)

bright_image = increase_brightness(image2)
cv2_imshow(image2)
cv2_imshow(bright_image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
